# Phương pháp

## Ảnh vệ tinh

*Khai thác sự khác biệt về phản xạ quang phổ giữa nước và các vật thể khác trên bề mặt Trái Đất*

* **Phân ngưỡng**: Sử dụng các chỉ số và cài giá trị ngưỡng để phân biệt nước:

$$ NDWI = \frac{GREEN - NIR}{GREEN + NIR}  \tag{1}$$

$$ mNDWI = \frac{GREEN - SWIR}{GREEN + SWIR}  \tag{2}$$

$$ AWEI_{nsh} = 4\times(GREEN SWIR1) - (0.25 \times NIR + 2.75 \times SWIR2) \tag{3.1}$$

$$ AWEI_{sh} = BLUE + 2.5\times{GREEN} - 1.5 \times (NIR + SWIR1) - 0.25 \times SWIR2 \tag{3.2}$$

⚙️Các bước trích xuất đường bờ:  
* Áp dụng các **chỉ số nước** lên hình ảnh (NDWI, mNDWI, AWEI, WI)
* Xử lý Otsu để **phân loại nhị phân** theo ngưỡng
* **Loại bỏ nhiễu, làm mịn** bằng lọc trung bình, Gaussian và *phép toán hình thái học*
* Trích xuất **biên** của vùng nước, vùng đất

* **Phân loại hình ảnh:** Sử dụng thuật toán phân loại:
    * Maximum Likelihood
    * SVM
    * Random Forest

> Tuy nhiên hạn chế lớn về yêu cầu dữ liệu
